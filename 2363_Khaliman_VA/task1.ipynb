## Основные теоретические положения

Функция плотности двумерного распределения: $p(\vec{x})=\frac{1}{2\pi\sqrt{\begin{vmatrix} Σ\end{vmatrix}}} exp\left(-\frac{1}{2}(\vec{x} - \vec{μ})^{\intercal}Σ^{-1}(\vec{x}-\vec{\mu})\right)$, где:

$\vec{x}= \begin{pmatrix} \xi \\ \eta\end{pmatrix}, \Sigma = \begin{pmatrix} \mathbb{D}\xi & \mathop{\mathrm{cov}}\limits(\xi, \eta) \\ \mathop{\mathrm{cov}}\limits(\eta, \xi) & \mathbb{D}\eta \end{pmatrix} $

$\vec{\mu} = \mathbb{E}\begin{pmatrix}\xi \\ \eta\end{pmatrix}$

Преобразования для пункта №2:
$\vec{\xi}_{st} \sim \mathcal{N}(\vec{\mu}_{st}, \Sigma_{st})$ — стандартный нормальный случайный вектор;\
$\vec{\xi}_{st} = A \cdot \vec{\xi} + b$ — афинное преобразование случайного вектора;\
$\vec{\mu}_{st} = A \cdot \vec{\mu} + b$ — получение стандартного нормального вектора математических ожиданий из исходного;\
$\Sigma_{st} = A \cdot \Sigma \cdot A^{\intercal}$ — получение стандартной нормальной ковариационной матрицы из исходной.

Формулы для пункта №3:
$\Sigma = Q \cdot D^{-1} \cdot Q^{\intercal}$

$p_{\xi,\eta}(x', y') = \frac{1}{2\pi \cdot \sqrt{|\Sigma|}} \cdot \exp\left(-\frac{1}{2} \cdot \left[  \begin{pmatrix} x' & y' \end{pmatrix} \cdot D^{-1} \cdot \begin{pmatrix} x' \\ y' \end{pmatrix} \right]\right)$

Условное распределение $\xi|\eta=x \sim \mathcal{N}(\mathbb{E}(\xi|\eta),\mathbb{D}(\xi|\eta))$.

### Плотность двумерного нормального распределения имеет вид: $p_{\xi,\eta} = C \cdot exp\left(-\frac{1}{2} (3x^2 - 3xy + 7y^2 + 9x + 8y + 13) \right)$;


#### 1) Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности $p_{\xi,\eta}$

$p_{\xi}(x_1, ... , x_n) = \frac{1}{(2\pi)^{n/2}\sqrt{detΣ}} ⋅ exp\left(-\frac{1}{2}(x-μ)⋅Σ^{-1}⋅(x-μ)^{\intercal} \right)$

Квадратичная форма:

$q(x,y) = 3x^2 - 3xy + 7y^2 + 9x + 8y + 13 = 3(x + 2)^2 - 3(x + 2)(y + 1) + 7(y + 1)^2$

$p_{\xi}(x_1, ... , x_n) = C \cdot exp\left(-\frac{1}{2} (3(x + 2)^2 - 3(x + 2)(y + 1) + 7(y + 1)^2) \right)$

Вектор математического ожидания:

$\mathbb{E}\begin{pmatrix}\xi \\ η \end{pmatrix} = \begin{pmatrix}-2 \\ -1 \end{pmatrix}$

Ковариационная матрица $Σ$:

$Σ^{-1} = \begin{pmatrix} 3 & -1.5 \\ -1.5 & 7\end{pmatrix}$

$\begin{vmatrix} Σ^{-1} \end{vmatrix} = \frac{75}{4}; \frac{1}{\begin{vmatrix} Σ^{-1} \end{vmatrix}} = \frac{4}{75}$

$Σ = \frac{4}{75} \cdot \begin{pmatrix} 7 & 1.5 \\ 1.5 & 3\end{pmatrix}$

Дисперсия:

$\mathbb{D}\xi = 7 \cdot \frac{4}{75} = \frac{28}{75}$

$\mathbb{D}\eta = 3 \cdot \frac{4}{75} = \frac{12}{75}$

Ковариация:

$\mathop{\mathrm{cov}}\limits(\xi, \eta) = \mathop{\mathrm{cov}}\limits(\eta, \xi) = 1.5 \cdot \frac{4}{75} = \frac{2}{25}$

Коэффициент корреляции:

$\rho(\xi, \eta) = \frac{\mathop{\mathrm{cov}}\limits(\xi, \eta)}{\sqrt{\mathbb{D}\xi \cdot \mathbb{D}\eta}} = \frac{\frac{2}{25}}{\sqrt{\frac{28}{75}\cdot\frac{12}{75}}} = \frac{\sqrt{21}}{14}$

Константа $C$:

$C = \frac{1}{2\pi\sqrt{\frac{1}{\begin{vmatrix} Σ^{-1} \end{vmatrix}}}} = \frac{1}{2π\sqrt{\frac{4}{75}}} = \frac{5\sqrt{3}}{4\pi}$

График плотности $p_{\xi,\eta}$:



```r

density_func <- function(x, y) {
  C <- (5 * sqrt(3)) / (4 * pi)
  exponent <- -0.5 * (3 * (x + 2)^2 - 3 * (x + 2) * (y + 1) + 7 * (y + 1)^2)
  return(C * exp(exponent))
}

x <- seq(-5, 5, length.out = 100)
y <- seq(-5, 5, length.out = 100)

density_matrix <- outer(x, y, density_func)
persp(x, y, density_matrix,
        xlab = "X-axis",
        ylab = "Y-axis",
        zlab = "Density",
        main = "График плотности",
        phi = 30, theta = 30)
```

$q(x,y) = 3(x + 2)^2 - 3(x + 2)(y + 1) + 7(y + 1)^2 = 3(x+2-0.5y-0.5)^2+6.25(y+1)^2 = \left(\sqrt{3}x-\frac{\sqrt{3}}{2}y+\frac{3\sqrt{3}}{2}\right)^2+\left(\frac{5}{2}y+\frac{5}{2}\right)^2$

Аффинное преобразование:

$Σ_{st}=\frac{4}{75} \cdot \begin{pmatrix} \sqrt{3} & -\frac{\sqrt{3}}{2} \\ 0 & \frac{5}{2}\end{pmatrix}\begin{pmatrix} 7 & 1.5 \\ 1.5 & 3\end{pmatrix}\begin{pmatrix} \sqrt{3} & 0 \\ -\frac{\sqrt{3}}{2} & \frac{5}{2}\end{pmatrix} = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}$

#### 3) Найти ортогональное преобразование, переводящее соответствующий центрированный случайный вектор в вектор с независимыми компонентами. Построить график плотности полученного распределения.

$Σ^{-1} = \begin{pmatrix} 3 & -1.5 \\ -1.5 & 7\end{pmatrix}$

$\begin{vmatrix} Σ^{-1}-λE\end{vmatrix} = 0$

$\begin{vmatrix} \begin{pmatrix} 3 & -1.5 \\ -1.5 & 7\end{pmatrix} - \begin{pmatrix} λ & 0 \\ 0 & λ\end{pmatrix} \end{vmatrix} = \begin{vmatrix}\begin{pmatrix}3-λ & -1.5 \\ -1.5 & 7-λ\end{pmatrix}\end{vmatrix} = 0$

$(3-λ)(7-λ)-2.25=0$

$\begin{cases}λ_1 = 7.5\\ λ_2 = 2.5\end{cases}$

$\begin{cases}-4.5n_1-1.5m_1=0 \\ -1.5n_1-0.5m_1=0\end{cases}$

$\begin{cases}n_1=-\frac{1}{3}m_1\\ m_1 = \alpha, \alpha∈\mathbb{R}\end{cases}$

$\alpha = \sqrt{\left(-\frac{1}{3}\right)^2+1^2} = \frac{\sqrt{10}}{3}$

$v_1 = \frac{3}{\sqrt{10}} \cdot \begin{pmatrix} -\frac{1}{3} \\ 1 \end{pmatrix}$

$\begin{cases}0.5n_2-1.5m_2=0 \\ -1.5n_2+4.5m_2=0\end{cases}$

$\begin{cases}n_2=3m_2 \\ m_2 = \beta, \beta \in \mathbb{R}\end{cases}$

$\beta = \sqrt{3^2+1^2} = \sqrt{10}$

$v_2 = \frac{1}{\sqrt{10}} \cdot \begin{pmatrix} 3 \\ 1 \end{pmatrix}$

$Q^{-1} = \frac{1}{\sqrt{10}}\begin{pmatrix} -1 & 3 \\ 3 & 1 \end{pmatrix} = Q$

$Q\cdotΣ^{-1}\cdot Q^{-1} = \frac{1}{\sqrt{10}}\begin{pmatrix} -1 & 3 \\ 3 & 1 \end{pmatrix} \cdot \begin{pmatrix} 3 & -1.5 \\ -1.5 & 7\end{pmatrix} \frac{1}{\sqrt{10}}\begin{pmatrix} -1 & 3 \\ 3 & 1 \end{pmatrix} = \begin{pmatrix} \frac{15}{2} & 0 \\ 0 & \frac{5}{2}\end{pmatrix} = \begin{pmatrix} 7.5 & 0 \\ 0 & 2.5\end{pmatrix} = A^{-1}$

$p_{\xi,\eta}(x', y') = \frac{1}{2\pi\sqrt{\begin{vmatrix}\Sigma\end{vmatrix}}} \cdot \exp\left(-\frac{1}{2} \cdot \left(  \begin{pmatrix} x' & y' \end{pmatrix} \cdot A^{-1} \cdot \begin{pmatrix} x' \\ y' \end{pmatrix} \right)\right) = \frac{5\sqrt{3}}{4\pi} \cdot\exp\left(-\frac{1}{2} \cdot ( 7.5x'^2 + 2.5y'^2)\right)$

```r
density_function <- function(x, y) {
  ((5 * sqrt(3)) / (4 * pi)) * exp(-0.5 * (7.5 * x^2 + 2.5 * y^2))
}

x_vals <- seq(-3, 3, length.out = 100)
y_vals <- seq(-3, 3, length.out = 100)
z_vals <- outer(x_vals, y_vals, density_function)

persp(x_vals, y_vals, z_vals, col = "lightblue",
        xlab = "X'", ylab = "Y'", zlab = "Density",
        main = "3D Density Distribution", phi = 30, theta = 30)
```

#### 4) Вычислить характеристики совместного распределения случайного вектора $(ξ+η, 2ξ+2η)^{\intercal}$ и записать его плотность. Построить график плотности полученного распределения.

$\vec{v} = A \cdot \begin{pmatrix} \xi \\ η \end{pmatrix} = (ξ+η, 2ξ+2η)^{\intercal}$

$A=\begin{pmatrix} 1 & 1 \\ 2 & 2\end{pmatrix}; \begin{vmatrix}A\end{vmatrix} = 0 \Rightarrow A=\begin{pmatrix} -1 & 1 \\ 2 & 2\end{pmatrix}$

$\vec{v} \sim \mathcal{N}(A\vec{μ}, A\cdot Σ \cdot A^{\intercal})$

$A \cdot \vec{μ} = \vec{μ}_{\vec{v}}; A\cdot Σ \cdot A^{\intercal} = Σ_{\vec{v}}$

$\vec{μ}_{\vec{v}} = \begin{pmatrix} -1 & 1 \\ 2 & 2\end{pmatrix} \cdot \begin{pmatrix} -2 \\ -1 \end{pmatrix} = \begin{pmatrix} 1 \\ -6 \end{pmatrix}$

$Σ_{\vec{v}} = \begin{pmatrix} -1 & 1 \\ 2 & 2\end{pmatrix} \cdot \frac{4}{75} \cdot \begin{pmatrix} 7 & 1.5 \\ 1.5 & 3\end{pmatrix} \cdot \begin{pmatrix} -1 & 2 \\ 1 & 2\end{pmatrix} = \frac{4}{75} \begin{pmatrix} 7 & -8 \\ -8 & 52 \end{pmatrix}$

$\mathbb{D}(\xi+\eta) = \frac{4}{75} \cdot 7 = \frac{28}{75}$

$\mathbb{D}(2\xi+2\eta) = \frac{4}{75} \cdot 52 = \frac{208}{75}$

$\mathop{\mathrm{cov}}\nolimits(\xi+\eta, 2\xi+2\eta) = \frac{4}{75}\cdot -8 = -\frac{32}{75}$

${\mathop{\mathrm{\rho}}\limits(\xi, \eta)} = \frac{\mathop{\mathrm{cov}}\limits(\xi + \eta, 2\xi + 2\eta)}{\sqrt{\mathbb{D}(\xi + \eta) \cdot \mathbb{D}(2\xi + 2\eta)}} = -\frac{4\sqrt{91}}{91}$

$\Sigma^{-1}_{\vec{v}} = \frac{1}{\begin{vmatrix}Σ_{\vec{v}}\end{vmatrix}}\cdot \begin{pmatrix} 52 & 8 \\ 8 & 7 \end{pmatrix} = \frac{4}{64}\cdot \begin{pmatrix} 52 & 8 \\ 8 & 7 \end{pmatrix}$

$p_{\vec{v}}(\vec{v})=\frac{1}{2\pi\sqrt{\frac{64}{75}}}\cdot \\ \cdot exp\left( -\frac{1}{2} \cdot \left( \begin{pmatrix} \xi + \eta \\ 2\xi + 2\eta \end{pmatrix} - \begin{pmatrix} 1 \\ -6 \end{pmatrix} \right)^{\intercal} \cdot \frac{4}{64}\cdot \begin{pmatrix} 52 & 8 \\ 8 & 7 \end{pmatrix} \cdot \left(\begin{pmatrix} \xi + \eta \\ 2\xi + 2\eta \end{pmatrix} - \begin{pmatrix} 1 \\ -6 \end{pmatrix} \right)\right) = \\= \frac{1}{2\pi\sqrt{\frac{64}{75}}}\cdot exp\left( -\frac{1}{2} \cdot \left( \begin{pmatrix} \xi + \eta - 1 \\ 2\xi + 2\eta +6\end{pmatrix} \right)^{\intercal} \cdot \frac{4}{64}\cdot \begin{pmatrix} 52 & 8 \\ 8 & 7 \end{pmatrix} \cdot \left(\begin{pmatrix} \xi + \eta -1 \\ 2\xi + 2\eta +6 \end{pmatrix}\right)\right)=\\ = \frac{1}{2\pi\sqrt{\frac{64}{75}}}\cdot exp\left( -\frac{1}{2} \cdot \frac{4}{64}\cdot (112x^2+224xy+112y^2+128x+128y+208)\right)$

```r
sigma2 <- 64 / 75
constant <- 1 / (2 * pi * sqrt(sigma2))

density_function <- function(x, y) {
  exponent <- -0.5 * (4 / 64) * (112 * (x + y)^2 + 128 * (x + y) + 208)
  return(constant * exp(exponent))
}

x_vals <- seq(-5, 5, length.out = 100)
y_vals <- seq(-5, 5, length.out = 100)
grid <- expand.grid(x = x_vals, y = y_vals)

grid$density <- mapply(density_function, grid$x, grid$y)

z_matrix <- matrix(grid$density, nrow = length(x_vals), ncol = length(y_vals))

persp(x_vals, y_vals, z_matrix,
      theta = 30, phi = 30, expand = 0.5,
      col = "lightblue", shade = 0.5,
      xlab = "X", ylab = "Y", zlab = "Плотность")
```
